#### Cohorts

```python
orders_= orders.copy()
orders_['min_date_buy'] =orders_.groupby(['Uid'])['Buy Ts'].transform('min')
orders_
```

```python
v_o = orders_.merge(first_visit, on='Uid', how='left')
v_o
```

```python
v_o['buy_week'] = pd.to_datetime(v_o['Buy Ts'].dt.date) - pd.to_timedelta(v_o['Buy Ts'].dt.dayofweek, unit='d')

v_o['min_visit_week'] = pd.to_datetime(v_o['min_start_ts'].dt.date) - pd.to_timedelta(v_o['min_start_ts'].dt.dayofweek, unit='d')
```

```python
v_o['cohort_lifetime'] = v_o['buy_week'] - v_o['min_visit_week']
v_o['cohort_lifetime'] = v_o['cohort_lifetime'] / np.timedelta64(1, 'W')
v_o['cohort_lifetime'] = v_o['cohort_lifetime'].astype(int)
v_o
```

```python
cohorts = (
        v_o.groupby(['buy_week', 'cohort_lifetime'], as_index=False)
        .agg({'Uid': 'nunique'})
        .sort_values(['buy_week', 'cohort_lifetime'])
    )

inital_users_count = cohorts[cohorts['cohort_lifetime'] == 0][
        ['buy_week', 'Uid']
    ]
inital_users_count = inital_users_count.rename(
        columns={'Uid': 'cohort_users'}
    )

cohorts = cohorts.merge(inital_users_count, on='buy_week')

cohorts['retention'] = cohorts['Uid'] / cohorts['cohort_users']

print(cohorts.groupby(['cohort_lifetime'])['retention'].mean())
cohorts.groupby(['cohort_lifetime'])['retention'].mean().plot.bar(rot=30, figsize=(12,5))
plt.show()
```

</div>


La mayoría de las compras ocurren en la semana de la primera visita. Aun así, un pequeño porcentaje de usuarios realiza compras recurrentes en las semanas siguientes. Estos datos confirman un patrón de conversión rápida pero retención limitada, típico en muchos entornos en linea. Sería estratégico fidelizar a quienes compran en semanas posteriores, ya que podrían representar un público más comprometido.

```python
cohort_v = v_o.groupby('min_visit_week')['Uid'].nunique().reset_index()
cohort_v.columns = ['cohort_date', 'unique_users']

cohort_b = v_o.groupby('buy_week')['Uid'].nunique().reset_index()
cohort_b.columns = ['cohort_date', 'unique_users']

cohort_compare = cohort_v.merge(cohort_b, on='cohort_date', how='outer').sort_values('cohort_date')


cohort_compare.columns=['cohort_date','visit_users', 'purchase_users']
cohort_compare.set_index('cohort_date')[['visit_users', 'purchase_users']].plot(
    kind='bar', figsize=(15,6), rot=45, color=['orange', 'blue']
)
plt.title('Usuarios únicos por semana: Visitas vs Compras')
plt.xlabel('Semana')
plt.ylabel('Cantidad de usuarios')
plt.tight_layout()
plt.show()
```

#### ¿Cuánto dinero traen? (LTV)

```python
purchase_
```

```python
orders_['Buy Ts'] = orders_['Buy Ts'].astype('datetime64[M]')
costs['dt'] = costs['dt'].astype('datetime64[M]')

first_orders = orders_.groupby('Uid').agg({'Buy Ts': 'min'}).reset_index()
first_orders.columns = ['Uid', 'first_order_month']
first_orders.head()

cohort_sizes = first_orders.groupby('first_order_month').agg({'Uid': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_order_month', 'n_buyers']
cohort_sizes.head()

orders_ = pd.merge(orders,first_orders, on='Uid')
orders_.head()

cohorts = orders_.groupby(['first_order_month','Buy Ts']).agg({'Revenue': 'sum'}).reset_index()
cohorts.head()

report = pd.merge(cohort_sizes, cohorts, on='first_order_month')
report.head()

margin_rate = 0.5

report['gp'] = report['Revenue'] * margin_rate
report['age'] = (
    report['Buy Ts'] - report['first_order_month']
) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')

report.head()

report['ltv'] = report['gp'] / report['n_buyers']

report

output = report.pivot_table(
    index='first_order_month', 
    columns='age', 
    values='ltv', 
    aggfunc='mean')

output.fillna('')
```

```python
plt.figure(figsize=(13,9))
plt.title ('Tamaño promedio de compra del cliente')
sns.set(style='white')
sns.heatmap(output,annot=True, fmt ='.4f', linewidths=1, linecolor= 'gray', vmin=0, vmax=0.005)
plt.show()
```

```python
ltv_por_cohorte = output.sum(axis=1)
ltv_por_cohorte=ltv_por_cohorte.reset_index()
ltv_por_cohorte.columns=['dt_m','ltv']
ltv_por_cohorte
```

```python
ltv_por_cohorte.mean().round(3)
```

El análisis del LTV (valor del tiempo de vida del cliente) por cohorte muestra una evolución significativa a lo largo del tiempo. Las cohortes más antiguas como la de septiembre de 2017  y agosto de 2017  presentan un LTV promedio más alto, lo que indica un comportamiento más rentable de esos usuarios en el tiempo. Por el contrario, las cohortes más recientes como febrero de 2018  y enero de 2018 tienen un LTV más bajo, lo que podría reflejar una menor retención o una menor venta promedio. La cohorte de junio de 2018 muestra un valor atípico (1.71) que probablemente correspondeuna venta aislada. En general, las cohortes de mediados de 2017 fueron las más valiosas, y se observa una tendencia decreciente del LTV en las cohortes más recientes, lo cual sugiere la necesidad de revisar la estrategia de retención o la calidad de los nuevos usuarios adquiridos.
